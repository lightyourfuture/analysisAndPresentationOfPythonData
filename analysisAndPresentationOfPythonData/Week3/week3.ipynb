#### pyplot 的绘图区域

`plt.subplot(nrows, ncols, plot_number)` 是 Matplotlib 库中的一个函数,用于在一个窗口中创建多个子图.这个函数非常有用,当你想在一个图形窗口中并排比较多个图表时.

##### 函数功能:

- `plt.subplot()` 用于在一个绘图窗口中创建一个子图.
- 子图会按照网格系统排列,网格的大小由 `nrows` 和 `ncols` 决定.
- `plot_number` 参数决定了当前子图的位置.

##### 输入参数说明:

1. **nrows**: 整数,表示子图网格的行数.
2. **ncols**: 整数,表示子图网格的列数.
3. **plot_number**: 整数,表示当前子图在网格中的位置.子图的编号从左到右,从上到下开始,从 1 开始编号.


````python
import numpy as np  # 导入numpy库,并将其别名为np
import matplotlib.pyplot as plt  # 导入matplotlib库的pyplot模块,并将其别名为plt


def f(t):  # 定义函数f,参数为t
    return np.exp(-t) * np.cos(2 * np.pi * t)  # 函数返回e的-t次方乘以cos(2πt)


a = np.arange(0.0, 5.0, 0.02)  # 使用numpy的arange函数生成一个从0.0到5.0,步长为0.02的数组,赋值给a

plt.subplot(211)  # 使用pyplot的subplot函数创建一个2行1列的图形,当前为第1个图

plt.plot(a, f(a))  # 使用pyplot的plot函数绘制折线图,x轴的值为a,y轴的值为f(a)

plt.subplot(2, 1, 2)  # 使用pyplot的subplot函数创建一个2行1列的图形,当前为第2个图

plt.plot(
    a, np.cos(2 * np.pi * a), "r--"
)  # 使用pyplot的plot函数绘制折线图,x轴的值为a,y轴的值为cos(2πa),线型为红色虚线

plt.show()  # 使用pyplot的show函数显示图形
````

### pyplot() 函数

`plt.plot()` 是 Matplotlib 库中用于绘制二维图形的基本函数.它可以绘制点和线.

#### 函数格式:

```python
plt.plot(x, y, format_string, **kwargs)
```

#### 参数说明:

- **x**: X 轴数据,列表或数组,可选.
- **y**: Y 轴数据,列表或数组.
- **format_string**: 控制曲线的格式字符串,可选.
- **kwargs**: 关键字参数,用于指定绘图属性,如线条颜色,宽度,标记等.

#### 注意事项:

- 当绘制多条曲线时,各条曲线的 `x` 不能省略.
- `format_string` 是可选的,它是控制曲线的颜色,标记和线型的简写方式.例如,'ro-' 表示红色实线,带圆圈标记.
- `**kwargs` 用于指定更多的属性,如线条宽度(`linewidth`), 标记大小(`markersize`) 等.


````python
import numpy as np  # 导入numpy库,并将其别名为np
import matplotlib.pyplot as plt  # 导入matplotlib库的pyplot模块,并将其别名为plt

a = np.arange(10)  # 使用numpy的arange函数生成一个从0到9的数组,赋值给a

plt.plot(
    a, a * 1.5, a, a * 2.5, a, a * 3.5, a, a * 4.5
)  # 使用pyplot的plot函数绘制四条线,x轴的值都为a,y轴的值分别为a乘以1.5,2.5,3.5和4.5

plt.show()  # 使用pyplot的show函数显示图形
````

#### format_string 参数

`format_string` 在 `plt.plot()` 函数中用于控制曲线的格式,包括颜色,风格和标记.这是一个可选参数,由颜色字符,风格字符和标记字符组成.

#### 颜色字符

颜色字符用于指定曲线的颜色.下表列出了一些常用的颜色字符及其说明:

| 颜色字符 | 说明          | 颜色字符 | 说明             |
| -------- | ------------- | -------- | ---------------- |
| `'b'`    | 蓝色 (blue)   | `'m'`    | 洋红色 (magenta) |
| `'g'`    | 绿色 (green)  | `'y'`    | 黄色 (yellow)    |
| `'r'`    | 红色 (red)    | `'k'`    | 黑色 (black)     |
| `'c'`    | 青绿色 (cyan) | `'w'`    | 白色 (white)     |

除了上述字符,你还可以使用十六进制颜色代码(如 `'#008000'`)或灰度值字符串(如 `'0.8'`)来指定颜色.

#### 风格字符

风格字符在 `plt.plot()` 函数的 `format_string` 参数中用于指定曲线的线型.下表列出了一些常用的风格字符及其说明:

| 风格字符        | 说明   |
| --------------- | ------ |
| `'-'`           | 实线   |
| `'--'`          | 破折线 |
| `'-.'`          | 点划线 |
| `':'`           | 虚线   |
| `''` (空字符串) | 无线条 |

风格字符允许你定义曲线的外观,使得图形更加清晰和区分.这些字符通常与颜色字符和标记字符组合使用,以提供多样化的曲线表示.

#### 标记字符

标记字符在 `plt.plot()` 函数的 `format_string` 参数中用于指定数据点的标记类型,例如圆圈,三角形等.这些标记有助于突出显示每个数据点.下表列出了一些常用的标记字符及其说明:

| 标记字符 | 说明             | 标记字符 | 说明         | 标记字符 | 说明         |
| -------- | ---------------- | -------- | ------------ | -------- | ------------ |
| `'.'`    | 点标记           | `'2'`    | 上花三角标记 | `'h'`    | 竖六边形标记 |
| `','`    | 像素标记(极小点) | `'3'`    | 左花三角标记 | `'H'`    | 横六边形标记 |
| `'o'`    | 实心圈标记       | `'4'`    | 右花三角标记 | `'+'`    | 十字标记     |
| `'v'`    | 下三角标记       | `'s'`    | 实心方形标记 | `'x'`    | x 标记       |
| `'^'`    | 上三角标记       | `'d'`    | 菱形标记     | `'D'`    | 瘦菱形标记   |
| `'<'`    | 左三角标记       | `'p'`    | 实心五角标记 | `'*'`    | 星形标记     |
| `'>'`    | 右三角标记       | `'P'`    | 星形标记     | `'_'`    | 水平线标记   |

标记字符通常与颜色字符和风格字符组合使用,以提供多样化的数据点表示.


````python
import numpy as np  # 导入numpy库,并将其别名为np
import matplotlib.pyplot as plt  # 导入matplotlib库的pyplot模块,并将其别名为plt

a = np.arange(10)  # 使用numpy的arange函数生成一个从0到9的数组,赋值给a

plt.plot(
    a,
    a * 1.5,
    "go-",  # 绘制第一条线,x轴的值为a,y轴的值为a乘以1.5,线型为绿色实线,标记为圆圈
    a,
    a * 2.5,
    "rx",  # 绘制第二条线,x轴的值为a,y轴的值为a乘以2.5,线型为无线,标记为红色x
    a,
    a * 3.5,
    "*",  # 绘制第三条线,x轴的值为a,y轴的值为a乘以3.5,线型为无线,标记为星号
    a,
    a * 4.5,
    "b-",  # 绘制第四条线,x轴的值为a,y轴的值为a乘以4.5,线型为蓝色实线,标记为无标记
)

plt.show()  # 使用pyplot的show函数显示图形
````

| 参数名            | 说明           | 示例                     |

### pyplot 的文本显示

Matplotlib 提供了一系列函数来添加文本和注解,这些功能可以增强图形的信息表达能力.以下是一些常用的函数及其说明:

| 函数             | 说明                     |
| ---------------- | ------------------------ |
| `plt.xlabel()`   | 对 x 轴增加文本标签      |
| `plt.ylabel()`   | 对 y 轴增加文本标签      |
| `plt.title()`    | 对图形整体增加文本标签   |
| `plt.text()`     | 在图形的任意位置增加文本 |
| `plt.annotate()` | 在图形中增加带箭头的注解 |

这些函数可以用来增加图形的描述性文本,使得图形的含义更加清晰易懂.


````python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个从0到5的等差数列,步长为0.02
a = np.arange(0.0, 5.0, 0.02)
# 创建一个正弦波图,"r--"表示红色虚线
plt.plot(a, np.cos(2 * np.pi * a), "r--")

# 设置x轴的标签,字体为SimHei,字体大小为15,颜色为绿色
plt.xlabel("横轴:时间", fontproperties="SimHei", fontsize=15, color="green")
# 设置y轴的标签,字体为SimHei,字体大小为15
plt.ylabel("纵轴:振幅", fontproperties="SimHei", fontsize=15)
# 设置图的标题,字体为SimHei,字体大小为25
plt.title(r"正弦波实例 $y=cos(2\pi x)$", fontproperties="SimHei", fontsize=25)
# 在图上添加文本,位置在(2,1),文本内容为"$\mu=100$",字体大小为15
plt.text(2, 1, r"$\mu=100$", fontsize=15)

# 设置x轴和y轴的显示范围
plt.axis([-1, 6, -2, 2])
# 显示网格
plt.grid(True)
# 显示图形
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个从0到5的等差数列,步长为0.02
a = np.arange(0.0, 5.0, 0.02)
# 创建一个正弦波图,"r--"表示红色虚线
plt.plot(a, np.cos(2 * np.pi * a), "r--")

# 设置x轴的标签,字体为SimHei,字体大小为15,颜色为绿色
plt.xlabel("横轴:时间", fontproperties="SimHei", fontsize=15, color="green")
# 设置y轴的标签,字体为SimHei,字体大小为15
plt.ylabel("纵轴:振幅", fontproperties="SimHei", fontsize=15)
# 设置图的标题,字体为SimHei,字体大小为25
plt.title(r"正弦波实例 $y=cos(2\pi x)$", fontproperties="SimHei", fontsize=25)
# 在图上添加注释,箭头指向(2,1),文本位置在(3,1.5),箭头的颜色为黑色,箭头的宽度为2
plt.annotate(
    r"$\mu=100$",
    xy=(2, 1),
    xytext=(3, 1.5),
    arrowprops=dict(facecolor="black", shrink=0.1, width=2),
)

# 设置x轴和y轴的显示范围
plt.axis([-1, 6, -2, 2])
# 显示网格
plt.grid(True)
# 显示图形
plt.show()
````

### pyplot饼图的绘制


````python
import matplotlib.pyplot as plt

# 定义每个饼图部分的大小
sizes = [15, 30, 45, 10]
# 定义每个部分的标签
labels = ["Frogs", "Hogs", "Dogs", "Logs"]
# 定义每个部分离中心的距离
explode = (0, 0.1, 0, 0)

# 创建饼图
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct="%1.1f%%",
    shadow=False,
    startangle=90,
)
plt.axis("equal")
# 显示饼图
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机数种子,以确保每次运行代码时生成的随机数都相同
np.random.seed(0)

# 定义正态分布的均值和标准差
mu, sigma = 100, 20

# 生成一个符合正态分布的随机数数组
a = np.random.normal(mu, sigma, size=100)

# 创建一个直方图,参数40表示将数据分为40个区间,histtype="stepfilled"表示使用阶梯形状的直方图,
# facecolor="b"设置直方图的颜色为蓝色,alpha=0.75设置直方图的透明度
plt.hist(a, 40, histtype="stepfilled", facecolor="b", alpha=0.75)

# 设置直方图的标题
plt.title("Histogram")

# 显示直方图
plt.show()
````

### pyplot 极坐标的绘制


````python
import numpy as np
import matplotlib.pyplot as plt

# 定义数据点的数量
N = 20
# 生成等差数列,范围是0到2π,用于表示极坐标的角度
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
# 生成N个0到10的随机数,用于表示极坐标的半径
radii = 10 * np.random.rand(N)
# 生成N个0到π/4的随机数,用于表示条形图的宽度
width = np.pi / 4 * np.random.rand(N)

# 创建一个极坐标子图
ax = plt.subplot(111, projection="polar")
# 在极坐标子图上创建条形图
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# 遍历每个条形图,设置颜色和透明度
for r, bar in zip(radii, bars):
    # 使用viridis颜色映射,颜色取决于半径
    bar.set_facecolor(plt.cm.viridis(r / 10.0))
    # 设置透明度
    bar.set_alpha(0.5)

# 显示图形
plt.show()
````

````python
import numpy as np  # 导入numpy库,并将其别名为np
import matplotlib.pyplot as plt  # 导入matplotlib库的pyplot模块,并将其别名为plt

N = 10  # 设置条形图的数量为10
theta = np.linspace(
    0.0, 2 * np.pi, N, endpoint=False
)  # 生成等差数列,范围从0到2π,共N个数,不包含终点值,作为每个条形的角度
radii = 10 * np.random.rand(N)  # 生成N个0到10之间的随机数,作为每个条形的半径
width = np.pi / 2 * np.random.rand(N)  # 生成N个0到π/2之间的随机数,作为每个条形的宽度

ax = plt.subplot(111, projection="polar")  # 创建一个极坐标子图
bars = ax.bar(theta, radii, width, bottom=0.0)  # 在极坐标子图上绘制条形图

for r, bar in zip(radii, bars):  # 遍历每个条形
    bar.set_facecolor(plt.cm.viridis(r / 10.0))  # 设置条形的颜色,使用viridis色彩映射,颜色取决于条形的半径
    bar.set_alpha(0.5)  # 设置条形的透明度为0.5

plt.show()  # 显示图形
````

## 实例二:引力波的绘制

引力波

- 物理学中,引力波是因为时空弯曲对外以辐射形式传播的能量.
- 爱因斯坦基于广义相对论预言了引力波的存在.
- 2016 年 6 月 16 日,LIG0 合作组宣布 2015 年 12 月 26 日 03:38:53(UTC),两台不同位置的引力波探测器同时探测到了一个引力波信号.
- 数据源：
- http://python123.io/dv/grawave.html
- http://python123.io/dv/H1_Strain.wav
- http://python123.io/dv/L1_Strain.wav
- http://python123.io/dv/wf_template.txt

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# 读取.wav音频文件
rate_h, hstrain = wavfile.read(
    r"H1_Strain.wav", "rb"
)  # 读取"H1_Strain.wav"文件,返回采样率和数据,赋值给rate_h和hstrain
rate_l, lstrain = wavfile.read(
    r"L1_Strain.wav", "rb"
)  # 读取"L1_Strain.wav"文件,返回采样率和数据,赋值给rate_l和lstrain

# 加载.txt模板文件
reftime, ref_H1 = np.genfromtxt(
    "GW150914_4_NR_waveform_template.txt"
).transpose()  # 使用numpy的genfromtxt函数读取.txt文件,并转置

# 计算时间向量
htime_interval = 1.0 / rate_h  # 计算H1音频的时间间隔
ltime_interval = 1.0 / rate_l  # 计算L1音频的时间间隔
htime_len = hstrain.shape[0] / rate_h  # 计算H1音频的时间长度
ltime_len = lstrain.shape[0] / rate_l  # 计算L1音频的时间长度
htime = np.arange(-htime_len / 2, htime_len / 2, htime_interval)  # 生成H1音频的时间向量
ltime = np.arange(-ltime_len / 2, ltime_len / 2, ltime_interval)  # 生成L1音频的时间向量

# 创建一个12x6的图形
fig = plt.figure(figsize=(12, 6))

# 在图形上创建一个2行2列的子图,当前为第1个图
plth = fig.add_subplot(221)
# 在子图上绘制H1 Strain的波形图,x轴的值为htime,y轴的值为hstrain,线型为黄色实线
plth.plot(htime, hstrain, "y")
# 设置x轴的标签为"Time (seconds)"
plth.set_xlabel("Time (seconds)")
# 设置y轴的标签为"H1 Strain"
plth.set_ylabel("H1 Strain")
# 设置子图的标题为"H1 Strain"
plth.set_title("H1 Strain")

# 在图形上创建一个2行2列的子图,当前为第2个图
pltl = fig.add_subplot(222)
# 在子图上绘制L1 Strain的波形图,x轴的值为ltime,y轴的值为lstrain,线型为绿色实线
pltl.plot(ltime, lstrain, "g")
# 设置x轴的标签为"Time (seconds)"
pltl.set_xlabel("Time (seconds)")
# 设置y轴的标签为"L1 Strain"
pltl.set_ylabel("L1 Strain")
# 设置子图的标题为"L1 Strain"
pltl.set_title("L1 Strain")

# 在图形上创建一个2行1列的子图,当前为第2个图
pltref = fig.add_subplot(212)
# 在子图上绘制模板的波形图,x轴的值为reftime,y轴的值为ref_H1
pltref.plot(reftime, ref_H1)
# 设置x轴的标签为"Time (seconds)"
pltref.set_xlabel("Time (seconds)")
# 设置y轴的标签为"Template Strain"
pltref.set_ylabel("Template Strain")
# 设置子图的标题为"Template"
pltref.set_title("Template")

# 调整子图的布局,使之填充整个图形区域
fig.tight_layout()
# 保存图形为"Gravitational_Waves_Original.png"
plt.savefig("Gravitational_Waves_Original.png")
# 显示图形
plt.show()
# 关闭图形,释放资源
plt.close(fig)
````
